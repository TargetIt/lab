# Number basics

```python
# prime
def isprime(num):
    '''
    Return True if num is a prime number, otherwise False
    '''   
    if num < 2:
        return False
    
    for i in range(2, int(math.sqrt(num))+1):
        if num % i == 0:
            return False
    return True
```

```python
# gcd
def gcd(a,b):
    while a != 0:
        a, b = b%a, a
        #print (a,b)
    return b

gcd(12,15)

#relatively - prime <==> gcd(a,b) ==1
```

```python
# lcm
def lcm(a,b):
    return abs(a*b)/gcd(a,b)

lcm(12,15)
```
